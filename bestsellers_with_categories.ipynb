# Exploratory Data Analysis (EDA)

```python
df.columns
```

```python
ax = sns.countplot(x = 'Genre', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

According to bars top bestsellers books are 'Non Fiction'.

```python
ax = sns.countplot(x = 'Year', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

According to bar graph we can see that bestsellers books are sold in 2009.

# Group By

```python
df.groupby(['Genre'], as_index=False)['User Rating'].sum().sort_values('User Rating', ascending=False)

sns.barplot(x = 'Genre' , y = 'User Rating' , data=df)
```

```python
df.groupby(['Genre'], as_index=False)['Reviews'].sum().sort_values('Reviews', ascending=False)

sns.barplot(x = 'Genre' , y = 'Reviews' , data=df)
```

According to bar graph we can se that bestsellers review books are 'Fiction' books.

```python
df.groupby(['Genre'], as_index=False)['Price'].sum().sort_values('Price', ascending=False)

sns.barplot(x = 'Genre' , y = 'Price' , data=df)
```

According to bar graph we can see that highest price books are 'Non Fiction' books.
