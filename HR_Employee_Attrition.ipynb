# Exploratory Data Analytics (EDA)

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender' , data=df)

for bars in ax.containers:
    ax.bar_label(bars)
    
```

Top employees are Male.

```python
ax = sns.countplot(x = 'Department' , data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

According to bar chart top department of the company is Research & Development Departmnet.

```python
ax = sns.countplot(x = 'EducationField' , data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

According to bar chart top most education field is Life Sciences.

# Group By

```python
df.groupby(['Department'], as_index=False)['MonthlyIncome'].sum().sort_values('MonthlyIncome', ascending=False)

sns.barplot(x='Department', y='MonthlyIncome', data=df)
```

According to graph highest monthly income department is Human Resources.

```python
df.groupby(['Department'], as_index=False)['MonthlyRate'].sum().sort_values('MonthlyRate', ascending=False)

sns.barplot(x='Department', y='MonthlyRate', data=df)
```

According to graph highest monthly rate department is Sales.

# Job Role

```python
df = df.groupby(['JobRole'], as_index=False)['MonthlyIncome'].sum().sort_values(by='MonthlyIncome', ascending=True)

plt.figure(figsize=(20,10))
sns.barplot(data=df, x='JobRole', y='MonthlyIncome')

plt.show()
```

Highest Monhtly income for the job role is Sales Executive.
